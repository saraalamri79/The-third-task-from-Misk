```python
import pandas as pd
import numpy as np
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
from surprise.model_selection import cross_validate
```

```python

reader = Reader(rating_scale=(1, 5))


data = Dataset.load_from_df(df[['user_id_maped', 'program_name', 'rating']], reader)
```

```python

trainset, testset = train_test_split(data, test_size=0.2, random_state=42)
```

```python

model = SVD()


model.fit(trainset)


predictions = model.test(testset)


rmse = accuracy.rmse(predictions)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
def get_top_n_recommendations(model, df, n=5):

    unique_users = df['user_id_maped'].unique()
    unique_programs = df['program_name'].unique()

    recommendations = {}

    for user in unique_users:

        predictions = [model.predict(user, program) for program in unique_programs]

        top_n = sorted(predictions, key=lambda x: x.est, reverse=True)[:n]


        recommendations[user] = [pred.iid for pred in top_n]

    return recommendations


top_recommendations = get_top_n_recommendations(model, df)

for user, programs in list(top_recommendations.items())[:5]:
    print(f"المستخدم {user}: {programs}")
```
